```python
import nltk
import heapq

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize


text = """India is a developing country. It has a large population. 
Technology is growing fast in India. Many people are using the internet. 
Education is also improving day by day. India is becoming digital."""

# Preprocessing
stop_words = set(stopwords.words('english'))
words = word_tokenize(text)

freq = {}
for word in words:
    word = word.lower()
    if word.isalnum() and word not in stop_words:
        freq[word] = freq.get(word, 0) + 1
        
max_freq = max(freq.values())

for word in freq:
    freq[word] = freq[word] / max_freq

sentences = sent_tokenize(text)
sent_score = {}

for sent in sentences:
    for word in word_tokenize(sent.lower()):
        if word in freq:
            sent_score[sent] = sent_score.get(sent, 0) + freq[word]

summary = heapq.nlargest(2, sent_score, key=sent_score.get)

print("Original Text:\n", text)
print("\nSummary:\n", " ".join(summary))
```
